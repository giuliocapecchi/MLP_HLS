# Visualize data inside the *'dataset1.txt* dataset

```python
import matplotlib.pyplot as plt

# open the file
with open('./linear_regression.txt', 'r') as f:
    # read the lines
    lines = f.readlines()
    # initialize the lists
    x = []
    y = []
    # iterate over the lines
    for line in lines:
        # split the line
        parts = line.split()
        # convert the parts to float and append them to the lists
        x.append(float(parts[0]))
        y.append(float(parts[1]))

# create the plot
plt.plot(x, y, 'ro')
# set the labels
plt.xlabel('x')
plt.ylabel('y')
# show the plot
plt.show();
```

```python
# open the ./final_weights.txt and get the slope and intercept
with open('./final_weights.txt', 'r') as f:
    m = float(f.readline())

# create the plot
plt.plot(x, y, 'ro')
# set the labels
plt.xlabel('x')
plt.ylabel('y')
# plot the line
plt.plot(x, [m * i for i in x])
# show the plot
plt.show();
```
